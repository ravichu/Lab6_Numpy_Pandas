```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/Most Popular Programming Languages.csv')
df['Month'] = pd.to_datetime(df['Month'])

print(df.head())
```

```python
student_id = 20250077
last3 = student_id % 1000
```

```python
languages = ['C#', 'Flutter', 'Java', 'JavaScript', 'Matlab', 'PhP', 'Python', 'React', 'Swift', 'TypeScript']
languages.sort()

index = last3 % len(languages)
assigned_language = languages[index]

print("Assigned Language:", assigned_language)
```

```python
lang_col = assigned_language + " Worldwide(%)"

data = df[['Month', lang_col]].copy()
data.rename(columns={lang_col: 'Popularity'}, inplace=True)

print(data.head())
```

```python
data['Growth_Rate'] = data['Popularity'].pct_change() * 100
```

```python
data['Moving_Avg'] = data['Popularity'].rolling(window=3).mean()
data['Moving_STD'] = data['Popularity'].rolling(window=3).std()
```

```python
conditions = [
    data['Growth_Rate'] > 5,
    data['Growth_Rate'] < -5
]

choices = ['Growth', 'Decline']

data['Phase'] = np.select(conditions, choices, default='Stable')
```

```python
summary = data['Popularity'].describe()
print(summary)
```

```python
phase_counts = data['Phase'].value_counts()
print(phase_counts)
```

```python
initial = data['Popularity'].iloc[0]
final = data['Popularity'].iloc[-1]

overall_growth = ((final - initial) / initial) * 100

print("Initial:", initial)
print("Final:", final)
print("Overall Growth %:", overall_growth)
```

```python
data['Growth_Rate'] = data['Popularity'].pct_change() * 100
```

```python
data['Moving_Avg'] = data['Popularity'].rolling(6).mean()
data['Moving_STD'] = data['Popularity'].rolling(6).std()
```

```python
mean_growth = np.mean(data['Growth_Rate'].dropna())
std_growth = np.std(data['Growth_Rate'].dropna())
```

```python
conditions = [
    (data['Growth_Rate'] > 0) & (data['Growth_Rate'] <= mean_growth),
    (data['Growth_Rate'] > mean_growth),
    (abs(data['Growth_Rate']) <= 1),
    (data['Growth_Rate'] < -std_growth)
]

choices = ['Introduction', 'Growth', 'Maturity', 'Decline']

data['Lifecycle'] = np.select(conditions, choices, default='Stable')
```

```python
counts = data['Lifecycle'].value_counts()
print(counts)
```
